```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
Options = Options()
Options.add_argument("--start-maximized") # 화면을 꽉 차게 띄우기
Options.add_experimental_option("detach", True)
```

# 1. 네이버로 이동

```python
browser = webdriver.Chrome(options=Options)

browser.get("http://naver.com")
```

# 2. 로그인 버튼 클릭

```python
elem = browser.find_element(By.CLASS_NAME, 'MyView-module__link_login___HpHMW')
elem.click()
```

# 3. id, pw 입력

```python
browser.find_element(By.ID, 'id').send_keys('input_id')
browser.find_element(By.ID, 'pw').send_keys('password')
```

# 4. 로그인 버튼 클릭

```python
browser.find_element(By.ID, 'log.login').click()
```

```python
time.sleep(3)
```

# 5. id를 새로 입력

```python
# browser.find_element(By.ID, 'id').send_keys('my_id')
browser.find_element(By.ID, 'id').clear() # 잘못된 Id 없애기
browser.find_element(By.ID, 'id').send_keys('my_id')
```

# 6. html 정보 출력

```python
print(browser.page_source)
```

# 7. 브라우조 종료

```python
time.sleep(3)
# browser.close() # 현재 탭만 닫음
browser.quit() # 전부 닫음
```
